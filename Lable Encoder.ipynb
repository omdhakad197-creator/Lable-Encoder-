#  Lable Encoder

```python
# jab koi order me data , (category data )ko  batana hota h tab use kare 
# low-0 ,medium-1, high-2
#high school , gradute, post g
#   order hai  - Label encoding use kare 
#   order nahi hai  - One Hot Encoder  ka use kare 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({"Name":["Amana","OM","shivangi","Ajay","MAAN","OM"]})
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit_transform(df["Name"])
```

```python
df["en_name"] = le.fit_transform(df["Name"])
```

```python
df
```

```python
new_data = df
new_data.to_csv("Label_Encoder.csv",index=False)
```
